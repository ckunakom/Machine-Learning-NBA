# ETL: Combine API data with data from the Kaggle CSV

We want the age for each player, which was not available in the API. We found a CSV on Kaggle that contained that data. In this notebook, we combined the two sets of data to get a csv with all desired data.

SOURCE: https://www.kaggle.com/justinas/nba-height-and-weight-analysis

```python
# Read in API stats data CSV
stats_df = pd.read_csv('data/allStatsFinalWithWinnerCol.csv')
stats_df.head()
```

```python
# Read in Kaggle CSV
ages_df = pd.read_csv('data/NBAPlayerAges1.csv')
ages_df.head()
```

```python
# Read in API game data CSV
games_df = pd.read_csv('data/allGamesFinalWithWinnerCol.csv')
games_df.head()
```

```python
# Merge dataframes
master_df = pd.merge(stats_df, games_df, how='left', on='gameID')
print(master_df.columns)
print(master_df.shape)
master_df.head()
```

```python
# Drop weird unnamed columns
master_df = master_df[['gameID', 'statsID', 'assists',
       'blocks', 'defensiveRebounds', 'threePointFieldGoalPercent',
       'threePointFieldGoalAttempted', 'threePointFieldGoal',
       'fieldGoalPercent', 'fieldGoalAttempt', 'fieldGoalMade',
       'freeThrowPercent', 'freeThrowAttempt', 'freeThrowMade', 'minutes',
       'offensiveRebound', 'personalFouls', 'playerID', 'playerFirstName',
       'playerHeightFeet', 'playerHeightInches', 'playerLastName',
       'playerPosition', 'team_id', 'playerWeightPounds', 'points', 'rebounds',
       'steals', 'teamID', 'turnovers', 'winner_x', 'date', 'homeTeamScore', 'visitorTeamScore',
       'season_y', 'time', 'homeTeamID', 'homeTeamAbbrev', 'homeTeamCity', 'homeTeamName', 'visitorTeamID',
       'visitorTeamAbbrev', 'visitorTeamCity', 'visitorTeamName']]

master_df.head()
```

```python
#master_df.to_csv('data/alitest2_112420.csv')
master_df = pd.read_csv('data/alitest2_112420.csv')
```

```python
print(games_df.columns)
print(stats_df.columns)
print(ages_df.columns)
print(master_df.columns)
```

```python
#ages_df.season.values[0][:4]
```

```python
## Take only the first year of season column to make merging easier
ages_df['season_y'] = ages_df['season_y'].map(lambda y: y[:4])
ages_df.head()
```

```python
# Rename column to make merging easier
ages_df = ages_df.rename(columns={'season': 'season_y'})
ages_df.columns
```

```python
# Add a column with both first and last name to make merging easier
master_df['playerFullName'] = master_df['playerFirstName'] + ' ' + master_df['playerLastName']
master_df.head()
```

```python
# change data types to make merging possible
master_df.dtypes
ages_df.dtypes
ages_df = ages_df.astype({'season_y': 'int64'})
```

```python
# merge on player name and the year to get AGES
master_final_df = pd.merge(master_df, ages_df, how='left', on=['playerFullName', 'season_y'])
```
